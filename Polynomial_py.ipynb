```python
# Author: 李子玉
# Date: [Date of Creation or Last Update]
# Purpose:

class Polynomial:
    """
    Class implementation of a polynomial. The polynomial coefficients are
    represented in a list in descending order of their degrees.
    """
    def __init__(self, coefficients):
        """
        Initialize the Polynomial instance.

        Args:
            coefficients (list): A list of coefficients, starting with the
                                 coefficient of the highest degree term.
        """
        # 建立屬性來儲存資料
        self._coeff = []
        self._degree = 0

        # 找尋第一個不是 0 的數字的位置，用來去除開頭的 0
        zeros_count = 0
        for c in coefficients:
            if c == 0:
                zeros_count += 1
            else:
                break

        # 處理如果輸入全都是 0 的情況
        if zeros_count == len(coefficients):
            self._coeff.append(0)
            self._degree = 0
        else:
            # 把從非 0 開始的剩餘元素，一個一個放進 self._coeff 裡 [cite: 46]
            for i in range(zeros_count, len(coefficients)):
                self._coeff.append(coefficients[i])

            # 算出這個多項式的最高次方數 (元素數量減 1)
            self._degree = len(self._coeff) - 1


    def __add__(self, other):
        """
        Add two polynomials.

        Args:
            other (Polynomial): Another Polynomial instance to add.

        Returns:
            Polynomial: The sum of the two polynomials.
        """
        # 第一步：看看這兩個多項式，誰比較長（誰的數字比較多）
        len_self = len(self._coeff)
        len_other = len(other._coeff)

        if len_self > len_other:
            max_len = len_self
        else:
            max_len = len_other

        # 第二步：準備一個夠大的新盒子，裡面先全部塞滿 0
        new_coeff = []
        for i in range(max_len):
            new_coeff.append(0)

        # 第三步：把「自己 (self)」的數字放進新盒子，記得要靠右對齊！
        # 算法：大盒子的長度 - 自己原本的長度 = 需要往右推幾格 (我們叫它位移量)
        offset_self = max_len - len_self
        for i in range(len_self):
            # 把數字加上去
            new_coeff[i + offset_self] = new_coeff[i + offset_self] + self._coeff[i]

        # 第四步：把「另一個 (other)」的數字也加進新盒子，一樣要靠右對齊！
        offset_other = max_len - len_other
        for i in range(len_other):
            # 把數字加上去
            new_coeff[i + offset_other] = new_coeff[i + offset_other] + other._coeff[i]

        # 第五步：把裝好答案的新盒子，做成一個新的「多項式 (Polynomial)」交出去
        return Polynomial(new_coeff)

    def __sub__(self, other):
        """
        Subtract one polynomial from another.

        Args:
            other (Polynomial): Another Polynomial instance to subtract.

        Returns:
            Polynomial: The result of the subtraction.
        """
        # 第一步：看看誰比較長
        len_self = len(self._coeff)
        len_other = len(other._coeff)

        if len_self > len_other:
            max_len = len_self
        else:
            max_len = len_other

        # 第二步：準備裝滿 0 的大盒子
        new_coeff = []
        for i in range(max_len):
            new_coeff.append(0)

        # 第三步：把「自己 (self)」靠右對齊放進去
        offset_self = max_len - len_self
        for i in range(len_self):
            new_coeff[i + offset_self] = new_coeff[i + offset_self] + self._coeff[i]

        # 第四步：把「另一個 (other)」靠右對齊「減」掉！ (注意這裡的減號)
        offset_other = max_len - len_other
        for i in range(len_other):
            new_coeff[i + offset_other] = new_coeff[i + offset_other] - other._coeff[i]

        # 第五步：自動呼叫 __init__ 幫我們處理前面可能變成 0 的數字
        return Polynomial(new_coeff)

    def __mul__(self, other):
        """
        Multiply two polynomials.

        Args:
            other (Polynomial): Another Polynomial instance to multiply with.

        Returns:
            Polynomial: The product of the two polynomials.
        """
        # 第一步：算出新盒子需要多大
        # 規律是：(左邊長度 + 右邊長度 - 1)
        len_self = len(self._coeff)
        len_other = len(other._coeff)
        new_len = len_self + len_other - 1

        # 第二步：準備一個裝滿 0 的新盒子
        new_coeff = []
        for i in range(new_len):
            new_coeff.append(0)

        # 第三步：雙重迴圈！讓兩邊的數字「每個配每個」互相乘起來
        for i in range(len_self):          # 從自己 (self) 依序挑一個數字
            for j in range(len_other):     # 從對方 (other) 依序挑一個數字

                # 把乘出來的結果，加到新盒子對應的位置 (i + j)
                new_coeff[i + j] = new_coeff[i + j] + (self._coeff[i] * other._coeff[j])

        # 第四步：一樣交給 __init__
        return Polynomial(new_coeff)

    def __neg__(self):
        """
        Negate the polynomial.

        Returns:
            Polynomial: A new Polynomial instance representing the negation of the current polynomial.
        """
        # 第一步：準備一個空的新盒子
        new_coeff = []

        # 第二步：把自己的數字一個一個拿出來，變號後放進新盒子
        for c in self._coeff:
            new_coeff.append(-c)

        # 第三步：把新盒子做成新的多項式交出去！
        return Polynomial(new_coeff)

    #### Do not modify the code below ####
    def __str__(self):
        """
        String representation of the polynomial.

        """
        terms = []
        first_term = True

        for i, coeff in enumerate(self._coeff):
            power = len(self._coeff) - i - 1
            if abs(coeff) > 1e-9:
                # Format coefficient and power
                coeff_str = "" if abs(coeff) == 1 and power != 0 else f"{coeff:.4f}".rstrip('0').rstrip('.')
                if power == 0:
                    term = coeff_str
                elif power == 1:
                    term = f"{coeff_str}x"
                else:
                    term = f"{coeff_str}x^{power}"
                # Add sign and format term
                if coeff > 0 and not first_term:
                    term = "+ " + term
                elif coeff < 0:
                    term = "- " + term.lstrip('-')
                terms.append(term)
                first_term = False

        return ' '.join(terms) if terms else "0"

    def __repr__(self):
        """
        Official string representation of the polynomial, for debugging.

        """
        return f"Polynomial({self._coeff})"
    #### Do not modify the code above ####
```
